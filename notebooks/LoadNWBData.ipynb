## Example of loading contents of NWB file and plotting data 

```python
from pynwb import NWBHDF5IO
from IPython.display import Markdown, display

import matplotlib.pyplot as plt

all_files = ['../NWB/LanoreEtAl2019.nwb']

for f in all_files:
    nwbfile_io = NWBHDF5IO(f, 'r')
    nwbfile = nwbfile_io.read()
    print('NWB file: %s'%f)
    fig, ax = plt.subplots(figsize=(12,5))
    
    for a in nwbfile.acquisition:
        #print("Looking at: %s"%a)
        
        if 'Control' in a: # prior to drug application
            
            d = nwbfile.acquisition[a]
            v = d.data[:]
            times = d.timestamps[:]
            print('Adding %i datapoints to plot'%(len(v)))
            plt.plot(times, v, label=a)
    
#plt.legend()
plt.show()
```
