## Loading NetPyNE simulation data 

```python
import matplotlib.pyplot as plt

# This will be produced by running: python HHTut_run.py 
netpyne_output = '../NetPyNE/HHTut/HHTut_data.json'

import json

fig, ax = plt.subplots(figsize=(12,5))
 
# Opening JSON file
with open(netpyne_output) as json_file:
    data = json.load(json_file)
    
    print('Loaded data containing: %s'%data.keys())
    
    for cell in data["simData"]["Vsoma"]: 
        print('Loading membrane potential of %s'%cell)
        v = data["simData"]["Vsoma"][cell]
        times = data["simData"]['t']
        
        
        plt.plot(times, v, label=cell)
 
plt.legend()
plt.show()
```
